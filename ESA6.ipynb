Es lohnt sich immer den Datensatz bereits vor der eigentlichen Verarbeitung anzuschauen. So erkennt man die Struktur und mögliche Fehler

```python
import pandas as pd
```

Laden wir nun den Datensatz(df) mit Pandas

```python
url = 'https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv'
df = pd.read_csv(url,index_col=0)
#df = pd.read_csv(url)
```

```python
df
```

```python
df.info()
```

Man sieht zwar bereits einige fehlende Werte, dennoch sollte dies bei großen Datensätzen nicht händisch oder Augenmaß erledigt werden

```python
df.isnull() 
```

Jetzt bereinigen wir den Datensatz um alle Zeilen mit vollständig fehlenden Werten

```python
df_drop = df.dropna(how='all') 
```

```python
df_drop
```

Deduplikation. Vollständig identische Zeilen werden gelöscht

```python
df_drop.duplicated(subset=None, keep='first')
```

```python
df_drop_dd = df_drop.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
df_drop_dd
```

Damit wäre die äußere Bereinigung abgeschlossen. Nun müssen verwertbare Werte aufbereitet werden.
Dazu gehören Jahreszahlen, Namen

```python
df_drop_dd['gender'].unique()
```

```python
mcv = df_drop_dd['gender'].value_counts()
```

```python
mcv
```

```python
mcv.keys()[0]
```

```python
df_drop_dd['gender'].fillna((mcv.keys()[0]), inplace=True)
```

Jetzt haben wir den NaN Wert im binären Feld 'gender' mit dem am häufigsten vorkommenden Wert beschrieben. Bei vielen fehlenden Werten wäre es besser das Verhältnis zwischen den Werten beizubehalten (male/female: 12/7). Auch sind diese Daten zur reinen Auswertung und nicht einer 1 zu 1 Verwendung geeignet.

```python
df_drop_dd
```

Nun sehen wir uns die Angaben des Alters an

```python
df_drop_dd['age'].value_counts()
```

Beim Alter treffen wir auf mehrere Probleme. Zum einen existieren offensichtliche Abweichungen wie String oder negative Werte. Zum Anderen haben wir Beispiele mit einem Alter von 2, 4 oder 6 Jahren mit einer eingetragenen E-Mail Adresse.

Als erstes bereinigen wir die offensichtlichen Werte. Die String Werte wie "old" und "yung" könnte man mit dataframe['age'].replace(["old", "yung"], ["65", "20"]) ersetzen. Anschließend sollten die Werte von "Object" in Int64 überführt werden. Da es viel mehr Cases geben könnte sparen wir uns den Schritt und setzen für fehlende Werte einen Durchschnittswert ein.

```python
 df_age = df_drop_dd['age'].replace(["old", "yung"], ["65", "20"])
```

```python
df_age
```

```python
df_drop_dd['age'] = df_age
```

```python
df_drop_dd['age'].dtypes
```
